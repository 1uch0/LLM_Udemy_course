```python
# Quantization Config - this allows us to load the model into memory and use less memory
#Reducing the precision is called Quantization. 32 bit number and replaxce it with 8 bit number. The accuracy does not decrease the accuracy a bit

quant_config = BitsAndBytesConfig( #Describe what type of quantization we want
    load_in_4bit=True,                 #We can put also 8but = true
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16, #Somre improvement in oerfiramnce
    bnb_4bit_quant_type="nf4"
)
```

```python
# Tokenizer

tokenizer = AutoTokenizer.from_pretrained(LLAMA) #We will crestr a tokenizer for Llama
tokenizer.pad_token = tokenizer.eos_token
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
```

```python
# The model, We need to load our modelº


model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)
```

```python
memory = model.get_memory_footprint() / 1e6
print(f"Memory footprint: {memory:,.1f} MB")
```

```python
outputs = model.generate(inputs, max_new_tokens=80)
print(tokenizer.decode(outputs[0]))
```

```python
# Clean up

del inputs, outputs, model
torch.cuda.empty_cache()
```

```python
# Wrapping everything in a function - and adding Streaming and generation prompts

def generate(model, messages):
  tokenizer = AutoTokenizer.from_pretrained(model)
  tokenizer.pad_token = tokenizer.eos_token
  inputs = tokenizer.apply_chat_template(messages, return_tensors="pt", add_generation_prompt=True).to("cuda") #Applied the chat template suitable for this toke izers
  streamer = TextStreamer(tokenizer)  #
  model = AutoModelForCausalLM.from_pretrained(model, device_map="auto", quantization_config=quant_config)
  outputs = model.generate(inputs, max_new_tokens=80, streamer=streamer) #up to 80 new tokens
  del tokenizer, streamer, model, inputs, outputs
  torch.cuda.empty_cache()
```

```python
generate(PHI3, messages)
```
