```python
##IMPORT LIBRARIES
import numpy as np # library in Python used for numerical computing. It provides support for large, multi-dimensional arrays and matrices
import matplotlib.pyplot as plt #creating static, animated, and interactive visualizations. It is widely used in data analysis,
import pandas as pd #Python library for data analysis and manipulation. It provides flexible data structures, primarily Series (1D arrays) and DataFrame (2D tables)
import gradio as gr

```

```python
#IMPORT THE DATASET

dataset = pd.read_csv('synthetic_cnc_machine_data_v3.csv')

print(dataset)
print("----------------Aqui es donde se importa el dataset organizado------------------")

X = dataset.iloc[:, 1:-1].values #Extract all the columns in the datse except the last one
y = dataset.iloc[:, -1].values #Usually in the datasets the last column is the dependet variable, output. In this case we dont want a range. Just the last column
print(X)
print(y)
```

```python
##Splitting data into test set and training set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
##FEATURE SCALING
#Standarization just for the training part of the data

#JUST APPLY FEAUTURE SCALING TO THE NUMERICAL VALUES, NOT TO THE DUMMY VARIABKLES (CLASSES), just aged and salary

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)


#We need to apply the same transforation in the same scaler to the test set

X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(sc.transform([[59.98, 5.35, 0.4]]))
print(y_pred)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
print(y_test)
```

```python
import gradio as gr
import matplotlib.pyplot as plt

# Keep your histories for plotting
temperature_history = []
energy_history = []
vibration_history = []
time_steps = []
counter = 0

def update_plot(temperature, energy_consumed, vibration_level):
    global counter
    counter += 1

    time_steps.append(counter)
    temperature_history.append(temperature)
    energy_history.append(energy_consumed)
    vibration_history.append(vibration_level)

    plt.figure(figsize=(8, 4))
    plt.plot(time_steps, temperature_history, label="Temperature (°C)", color='r')
    plt.plot(time_steps, energy_history, label="Energy Consumed (kWh)", color='g')
    plt.plot(time_steps, vibration_history, label="Vibration Level", color='b')
    plt.xlabel("Time Step")
    plt.ylabel("Value")
    plt.title("Machine Parameters Over Time")
    plt.legend()
    plt.tight_layout()
    return plt

def predict_failure(temperature, energy_consumed, vibration_level):
    y_pred = classifier.predict(sc.transform([[temperature, energy_consumed, vibration_level]]))
    return "⚠️ Machine is FAILING!" if y_pred[0] == 1 else "✅ Machine is OK."

# Simple "machine scoring" for altruistic collaboration
def evaluate_machine(temp, energy, vibration, load, wear, availability):
    # This is just an example score function → you can replace with your own logic
    health = max(0, 100 - (load + wear*50 + vibration*100)) * availability
    return f"Score: {round(health,2)}"

with gr.Blocks() as demo:
    gr.Markdown("## 🏭 Machine Monitoring & Altruistic Collaboration")

    with gr.Tabs():
        # -------- Tab 1: Single machine --------
        with gr.Tab("Single Machine"):
            gr.Markdown("### Machine Parameters")

            with gr.Row():
                temp = gr.Slider(0, 100, 59.98, step=1, label="Temperature (°C)")
                energy = gr.Slider(0, 10, 5.35, step=0.1, label="Energy Consumed (kWh)")
                vibration = gr.Slider(0, 1, 0.4, step=0.01, label="Vibration Level")

            with gr.Row():
                tool_load = gr.Slider(0, 100, 30, step=1, label="Tool Load (%)")
                tool_wear = gr.Slider(0, 1, 0.2, step=0.01, label="Tool Wear (0-1)")
                availability = gr.Slider(0, 1, 0.9, step=0.01, label="Availability")

            plot_output = gr.Plot()
            status_output = gr.Textbox(label="Machine Status", interactive=False)
            score_output = gr.Textbox(label="Altruistic Collaboration Score", interactive=False)

            # Live plot updates
            temp.change(update_plot, [temp, energy, vibration], plot_output)
            energy.change(update_plot, [temp, energy, vibration], plot_output)
            vibration.change(update_plot, [temp, energy, vibration], plot_output)

            # Prediction button
            predict_btn = gr.Button("Predict Machine Failure")
            predict_btn.click(predict_failure, [temp, energy, vibration], status_output)

            # Collaboration score
            collab_btn = gr.Button("Evaluate Machine Collaboration Potential")
            collab_btn.click(evaluate_machine, [temp, energy, vibration, tool_load, tool_wear, availability], score_output)

        # -------- Tab 2: Compare 4 machines --------
        with gr.Tab("Compare 4 Machines"):
            gr.Markdown("### Multi-Machine Comparison for Altruistic Collaboration")

            machines = []
            for i in range(1,5):
                with gr.Column():
                    gr.Markdown(f"**Machine {i}**")
                    t = gr.Slider(0, 100, 50, step=1, label=f"M{i} Temperature (°C)")
                    e = gr.Slider(0, 10, 5, step=0.1, label=f"M{i} Energy (kWh)")
                    v = gr.Slider(0, 1, 0.3, step=0.01, label=f"M{i} Vibration")
                    l = gr.Slider(0, 100, 30, step=1, label=f"M{i} Tool Load (%)")
                    w = gr.Slider(0, 1, 0.2, step=0.01, label=f"M{i} Tool Wear (0-1)")
                    a = gr.Slider(0, 1, 0.9, step=0.01, label=f"M{i} Availability")
                    score_box = gr.Textbox(label=f"M{i} Score", interactive=False)
                    machines.append(((t,e,v,l,w,a), score_box))

            # One button to evaluate all 4 machines
            compare_btn = gr.Button("Evaluate All Machines")
            for (inputs, out) in machines:
                compare_btn.click(evaluate_machine, list(inputs), out)

demo.launch()
```
